```python
from surprise import Reader
from surprise import accuracy
from surprise.model_selection import train_test_split
```

```python
data1=pd.read_csv('Amazon - Movies and TV Ratings.csv')
```

```python
#getting column names

data1.columns
```

```python
melt_data=data1.melt(id_vars=data1.columns[0],value_vars=data1.columns[1:],var_name="movie_name",value_name="rating")
```

```python
melt_data
```

```python
reader=Reader(rating_scale=(-1,10))
```

```python
from surprise import Dataset
```

```python
df=Dataset.load_from_df(melt_data.fillna(0),reader=reader)
```

```python
#Splitting data into training and test set

trainset,testset=train_test_split(df,test_size=0.2)
```

```python
from surprise import SVD
from surprise.model_selection import cross_validate
```

```python
algo=SVD()
```

```python
#Training model

algo.fit(trainset)
```

```python
#Predicting test value

pred=algo.test(testset)
```

```python
#Obtaining performance metric

print(accuracy.rmse(pred))
```

```python
from surprise.model_selection import GridSearchCV
```

```python
param_grid={'n_epochs':[20,30],
           'lr_all':[0.005,0.01],
           'n_factors':[50,100]}
```

```python
grid_search=GridSearchCV(SVD,param_grid,measures=['rmse','mae'],cv=3,n_jobs=1)
```

```python
grid_search.fit(df)
```
